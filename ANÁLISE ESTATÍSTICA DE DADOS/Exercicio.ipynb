```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel('ibge.xlsx', 'Tab 1.1.38')

# Definindo os nomes das colunas
colNm = [
    'Regiao',
    'MedioTotal',
    'MedioUtilizaram',
    'MedioNaoUtilizaram',
    'MedianoTotal',
    'MedianoUtilizaram',
    'MedianoNaoUtilizaram',
]

# Aplicando os nomes das colunas ao DataFrame
df.columns = colNm

# Removendo as linhas específicas
df = df.drop([0, 1, 2, 3, 46, 47])

# Definindo a coluna 'Regiao' como índice
df = df.set_index('Regiao')

# Removendo a linha correspondente ao 'Brasil'
df = df.drop(['Brasil'])

# Ordenando o DataFrame pela coluna 'MedioTotal'
df = df.sort_values('MedioTotal')

# Convertendo os valores para numéricos, ignorando erros
df = df.apply(pd.to_numeric, errors='coerce')

# Criando o gráfico de linhas com rotação de 90 graus nos rótulos
df.plot(rot=90)

# Gerando a matriz de correlação
matCorr = df.corr()

# Limpando quaisquer figuras anteriores
plt.gcf().clear()  # Use gcf().clear() para limpar a figura atual

# Criando o gráfico de caixa (boxplot) com rotação de 90 graus nos rótulos
df.boxplot(rot=90)

# Mostrando o gráfico
plt.show()
```
