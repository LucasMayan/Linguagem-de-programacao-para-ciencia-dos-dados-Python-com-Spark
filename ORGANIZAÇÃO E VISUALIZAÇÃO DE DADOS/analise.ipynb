```python
import pandas as pd


df = pd.read_excel('ibge.xlsx', 'Tab 1.1.38')
```

```python

colNm = [
         'Regiao',
          'rem1',
          'utilizaramMensalPCapita',
          'NaoUtilizaramMensalPCapita',
          'rem2',
          'rem3',
          'rem4',
          ]

df.columns = colNm

df = df.drop([0, 1 ,2 ,3 , 46, 47])

df = df.drop(columns=['rem1', 'rem2', 'rem3', 'rem4'])

df = df.set_index(df['Regiao'])

df = df.drop(columns=['Regiao'])

df['utilizaramMensalPCapita'].mean()
df['NaoUtilizaramMensalPCapita'].mean()

df.plot.line(rot=90, figsize=(10,3))

df['utilizaramMensalPCapita'] = pd.to_numeric(df['utilizaramMensalPCapita'])
df['NaoUtilizaramMensalPCapita'] = pd.to_numeric(df['NaoUtilizaramMensalPCapita'])

df.hist('utilizaramMensalPCapita')
df.hist('NaoUtilizaramMensalPCapita')
```
